```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

no_dox = pd.read_csv('D:\\wound_assay\\no_dox.csv').to_numpy().flatten()
plus_dox = pd.read_csv('D:\\wound_assay\\plus_dox.csv').to_numpy().flatten()

no_dox = np.where(no_dox<0, no_dox+360, no_dox)
plus_dox = np.where(plus_dox<0, plus_dox+360, plus_dox)


# violinplot = plt.violinplot([no_dox,plus_dox], showmedians=True, points=200, bw_method=0.4, widths=0.7, showextrema=False)
# violinplot['cmedians'].set_color('black')
# for violin, color in zip(violinplot['bodies'], ['lightgrey','cornflowerblue','lightsalmon']):
#     violin.set_facecolor(color)
#     violin.set_edgecolor('black')
#     violin.set_alpha(1)
# plt.show() 

violinplot = plt.violinplot([no_dox,plus_dox], showextrema=False, showmedians=False, widths=0.9, points=200, bw_method=0.3)
# violinplot['cmedians'].set_color('black')
for violin, color in zip(violinplot['bodies'], ['lightgreen','orchid']):
    violin.set_facecolor(color)
    violin.set_edgecolor('black')
    violin.set_alpha(0.8)
boxplot = plt.boxplot([no_dox,plus_dox], whis=(0,100), showfliers=True, medianprops={'color':'black'}, widths=0.2, capwidths=0.1)
plt.yticks([0,30,60,90,120,150,180,210,240,270,300,330,360])
plt.ylim(0,400)

from scipy.stats import ranksums
ranksums(no_dox,plus_dox)
```
